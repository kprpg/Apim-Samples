### 🛠️ Configure Infrastructure Parameters & Create the Infrastructure

Set your desired parameters for the Simple APIM infrastructure deployment.

❗️ **Modify entries under _User-defined parameters_**.

```python
from pathlib import Path
import os
import sys
from apimtypes import APIM_SKU, INFRASTRUCTURE, Region
from utils import InfrastructureNotebookHelper
from console import print_ok

workspace_root = Path.cwd()
project_root = None
for candidate in [workspace_root, *workspace_root.parents]:
    if (candidate / 'shared' / 'python').exists():
        project_root = candidate
        break

if project_root is None:
    raise ModuleNotFoundError("Could not locate 'shared/python' from current notebook path")

shared_python_path = project_root / 'shared' / 'python'
for path_entry in (shared_python_path, project_root):
    path_value = str(path_entry)
    if path_value not in sys.path:
        sys.path.insert(0, path_value)

py_path_parts = [str(shared_python_path), str(project_root)]
existing_py_path = os.environ.get('PYTHONPATH', '')
if existing_py_path:
    py_path_parts.append(existing_py_path)
os.environ['PYTHONPATH'] = os.pathsep.join(py_path_parts)



# ------------------------------
#    USER CONFIGURATION
# ------------------------------

rg_location = Region.EAST_US_2      # Azure region for deployment
index       = 1                     # Infrastructure index (use different numbers for multiple environments)
apim_sku    = APIM_SKU.BASICV2      # Options: 'DEVELOPER', 'BASIC', 'STANDARD', 'PREMIUM', 'BASICV2', 'STANDARDV2', 'PREMIUMV2'

# ------------------------------
#    SYSTEM CONFIGURATION
# ------------------------------

inb_helper = InfrastructureNotebookHelper(rg_location, INFRASTRUCTURE.SIMPLE_APIM, index, apim_sku)
inb_helper.create_infrastructure()

print_ok('All done!')
```
