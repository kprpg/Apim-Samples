### 🗑️ Clean up resources

When you're finished with the lab, you should remove all your deployed resources from Azure to avoid extra charges and keep your Azure subscription uncluttered.

```python
from pathlib import Path
import os
import sys

workspace_root = Path.cwd()
project_root = None
for candidate in [workspace_root, *workspace_root.parents]:
    if (candidate / 'shared' / 'python').exists():
        project_root = candidate
        break

if project_root is None:
    raise ModuleNotFoundError("Could not locate 'shared/python' from current notebook path")

shared_python_path = project_root / 'shared' / 'python'
for path_entry in (shared_python_path, project_root):
    path_value = str(path_entry)
    if path_value not in sys.path:
        sys.path.insert(0, path_value)

py_path_parts = [str(shared_python_path), str(project_root)]
existing_py_path = os.environ.get('PYTHONPATH', '')
if existing_py_path:
    py_path_parts.append(existing_py_path)
os.environ['PYTHONPATH'] = os.pathsep.join(py_path_parts)

from apimtypes import INFRASTRUCTURE
from infrastructures import cleanup_infra_deployments

indexes = [1]

cleanup_infra_deployments(INFRASTRUCTURE.AFD_APIM_PE, indexes)
```
